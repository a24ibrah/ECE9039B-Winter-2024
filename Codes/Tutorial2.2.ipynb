The tutorial below demonstrates how to create a simple linear classification model (AKA Perceptron) using Python and the scikit-learn library. It also includes the performance metrics we discussed in class.

Step 1: Import necessary libraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split

import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2: Load the dataset

```python
data = load_breast_cancer()
X = data.data
y = data.target

# Make the target labels binary (+1, -1)
y_binary = [1 if label == 1 else -1 for label in y]
```

Step 3: Split the dataset into training and testing set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)
```

Step 4: Create and train the Perceptron model (basic linear classifier)

```python
classifier = Perceptron(tol=1e-3, random_state=42)
classifier.fit(X_train, y_train)
```

P.S. the parameter "tol=1e-3" refers to the tolerance for stopping criteria of the Perceptron algorithm. 

Step 5: Make predictions on the test set

```python
y_pred = classifier.predict(X_test)
```

Step 6: Evaluate the model using various metrics and display the values!

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label=1)
recall = recall_score(y_test, y_pred, pos_label=1)
f1 = f1_score(y_test, y_pred, pos_label=1)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

Step 7: Display the Confusion Matrix with values and visualize it

```python
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues')
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix for Perceptron Model')
plt.show()
```
